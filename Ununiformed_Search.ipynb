```python
graph = {
    'S':['A','B'],
    'A':['C','D'],
    'B':['A','D'],
    'C':['G'],
    'D':['G'],
    'G':[]
}
```

```python
from collections import deque

def bfs(graph, start,goal):
    visited = set()
    queue = deque([start])
    parent = {start: None}
    max_frontier = 0
    
    print("BFS: ")
    while queue:
        max_frontier = max(max_frontier, len(queue))
        node = queue.popleft()
        
        if node not in visited:
            print ( node, end=" ")
            visited.add(node)
            if node == goal: break
            
            for neighbor in sorted(graph[node]):
                if neighbor not in visited and neighbor not in queue:
                    parent[neighbor] = node
                    queue.append(neighbor)

    print(f"\nMax Frontier (Queue) size: {max_frontier}")
    path = []
    curr = goal
    while curr:
        path.append(curr)
        curr = parent.get(curr)
    return path[::-1], max_frontier


def dfs(graph, start,goal):
    visited = set()
    stack = deque([start])
    parent = {start: None}
    max_depth = 0
    
    print("DFS: ")
    while stack:
        max_stack_depth = max(max_depth, len(stack))
        node = stack.pop()
        
        if node not in visited:
            print ( node, end=" ")
            visited.add(node)

            if node == goal: break
            
            for neighbor in sorted(graph[node], reverse=True):
                if neighbor not in visited:
                    parent[neighbor] = node
                    stack.append(neighbor)
    print(f"\nMax Stack Depth reached: {max_stack_depth}")

    path = []
    curr = goal
    while curr:
        path.append(curr)
        curr = parent.get(curr)
    return path[::-1], max_depth
```

```python
bfs_path, bfs_max_q = bfs(graph, 'S', 'G')
print("\n")
dfs_path, dfs_max_s = dfs(graph, 'S', 'G')

print(f"\n\nBFS Path: {bfs_path}")
print(f"DFS Path: {dfs_path}")
```
