```python
graph = {
    'A':['B','C'],
    'B':['D','E'],
    'C':['F'],
    'D':[],
    'E':[],
    'F':[]
}
```

```python
from collections import deque
```

```python
def dfs(graph, start):
    visited = set()
    stack = deque([start])

    while stack:
        node = stack.pop()
        if node not in visited:
            print (node, end=" ")
            visited.add(node)
            stack.extend(graph[node]) #pop from right
            stack.extend(graph[node][::-1]) #Pop from left
```

```python
#Call DFS
dfs(graph,'A')
```
