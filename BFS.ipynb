```python
graph = {
    'A':['B','C'],
    'B':['D','E'],
    'C':['F'],
    'D':[],
    'E':[],
    'F':[]
}
```

```python
from collections import deque
```

```python
def bfs(graph, start):
    visited = set()
    queue = deque([start])

    while queue:
        node = queue.popleft()
        if node not in visited:
            print (node, end=" ")
            visited.add(node)
            queue.extend(graph[node])
```

```python
#Call BFS
bfs(graph,'A')
```
